```python
import os

import numpy as np
import skimage as ski
from matplotlib import pyplot as plt
```

```python
model_path = "../../../output/tandem_160_white/results"
img_name = "00004.png"
```

```python
gt_path = os.path.join(model_path, "unpolarized", "test", "ours_30000", "gt", img_name)
unpolarized_path = os.path.join(model_path, "unpolarized", "test", "ours_30000", "renders", img_name)
composite_path = os.path.join(model_path, "composite", "test", "ours_30000", "renders", img_name)
direct_path = os.path.join(model_path, "direct", "test", "ours_30000", "renders", img_name)
global_path = os.path.join(model_path, "global", "test", "ours_30000", "renders", img_name)

gt_img = ski.io.imread(gt_path)
unpolarized_img = ski.io.imread(unpolarized_path)
composite_img = ski.io.imread(composite_path)
direct_img = ski.io.imread(direct_path)
global_img = ski.io.imread(global_path)
```

```python
fig = plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(gt_img)
plt.title("Ground Truth")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(unpolarized_img)
plt.title("Baseline")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(composite_img)
plt.title("Direct-Global Separated")
plt.axis('off')

# plt.subplot(1, 5, 4)
# plt.imshow(direct_img)
# plt.title("Direct")
# plt.axis('off')

# plt.subplot(1, 5, 5)
# plt.imshow(global_img)
# plt.title("Global")
# plt.axis('off')

# plt.tight_layout()
plt.show()
fig.savefig("comparison.svg", bbox_inches='tight', dpi=300)
```
